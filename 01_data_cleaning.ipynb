#### Data Cleaning

```python
df = pd.read_csv("amazon.csv")
```

```python
df.dropna(subset = ['rating_count'], axis=0, inplace=True)
```

```python
df.isna().sum()
```

```python
# remove unwanted values and change dtype for better analysis
cols_to_float=["discounted_price","actual_price"]
df[cols_to_float]=df[cols_to_float].replace(r"₹|,","",regex=True)
df[cols_to_float]=df[cols_to_float].astype("float")
df["discount_percentage"]=df["discount_percentage"].str.replace("%","").astype("int")
df["rating_count"]=df["rating_count"].str.replace(",","").astype("int")
```

```python
# handling the valueerror - could not convert string to float: '|'
display(df[df["rating"].str.contains(r"\|")])
```

```python
# remove unwanted values and change dtype of rating
df["rating"] = np.where(df["rating"]=="|",np.nan,df["rating"])
df.dropna(axis=0,inplace=True)
df["rating"]=df["rating"].astype("float")
```

```python
df.isna().sum()
```

```python
# splitting category column into multiple columns for clean data
new_cols = df['category'].str.split('|', expand=True)
df['main_category'] = new_cols[0]
df['sub_category'] = new_cols[1]
df['sub_sub_category'] = new_cols[2]
df['level_4'] = new_cols[3]
df['level_5'] = new_cols[4]
df['level_6'] = new_cols[5]
df['level_7'] = new_cols[6]
```

```python
# Manually delete the temp variable to free up RAM
del new_cols
```

```python
# drop the non required column
df.drop(columns='category', inplace=True)
```

```python
# To reorder columns
# Define your "Front" columns
front_cols = ['product_id', 'product_name', 'main_category', 'sub_category']

# Get the rest of the columns automatically
remaining_cols = [c for c in df.columns if c not in front_cols]

# Combine and reorder
df = df[front_cols + remaining_cols]
```

```python
df.to_csv("cleaned_amazon.csv", index=False)
```
